```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def plotting(df):
    plt.figure(figsize=(10, 6))
    plt.plot(df['time'], df['ax'], label='ax')
    plt.plot(df['time'], df['ay'], label='ay')
    plt.plot(df['time'], df['az'], label='az')
    plt.xlabel('Time')
    plt.ylabel('Acceleration')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
file_name='Walking_4.csv'
df = pd.read_csv(os.path.join('Csv files',file_name),sep=",")
print(df)
```

```python
plotting(df)
```

```python
trim_point=8
trimmed_df = df[df['time'] >= trim_point]
plotting(trimmed_df)
```

```python
trimmed_df.shape
```

```python
downsampled_df = trimmed_df.iloc[::12, :]
```

```python
downsampled_df.shape
```

```python
plotting(downsampled_df)
```

```python
segmented_df = downsampled_df.iloc[:500, :]
segmented_df.shape
```

```python
plotting(segmented_df)
```

```python
final_df = segmented_df[['ax', 'ay', 'az']]
final_df.shape
```

```python
final_df
```

```python
final_df.to_csv(os.path.join('Processed files',file_name),sep=",",index=False)
```
