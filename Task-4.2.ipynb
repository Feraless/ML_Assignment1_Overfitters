```python
df = pd.DataFrame()
df_test = pd.DataFrame()
dict1 = {"LAYING":0,
"SITTING":1,
"STANDING":2,
"WALKING":3,
"WALKING_DOWNSTAIRS":4,
"WALKING_UPSTAIRS":5}
```

```python

base_path = "/Users/tanishqchaudhari/Desktop/Cloned Git Repos/ML_Assignment_1/human+activity+recognition+using+smartphones/UCI HAR Dataset/Combined/Train"

# Initialize an empty DataFrame
df = pd.DataFrame()

for files in os.listdir(base_path):
    full_path = os.path.join(base_path, files)
    if os.path.isdir(full_path):  # Ensure `files` is a directory
        for dfs in os.listdir(full_path):
            dfs_full_path = os.path.join(full_path, dfs)
            if os.path.isfile(dfs_full_path):  # Ensure `dfs` is a file
                df1 = pd.read_csv(dfs_full_path)
                df1["Label"] = np.array([dict1[files]] * df1.shape[0])
                df = pd.concat([df, df1], axis=0)

# Now `df` contains the combined DataFrame
```

```python
dt = DecisionTreeClassifier(criterion="gini", random_state=42,
        splitter="best",
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        min_weight_fraction_leaf=0.0,
        max_features=None,
        max_leaf_nodes=None,
        min_impurity_decrease=0.0,
        class_weight=None,
        ccp_alpha=0.0,
        monotonic_cst=None)
```

```python
dt.fit(X_train_tsfel[1:126], y_train_tsfel)
```

```python
predictions = dt.predict(X_test_tsfel[1:25])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
accuracy = accuracy_score(predictions, y_test_tsfel)
print(f"Accuracy: {accuracy}")
```

```python
precision = precision_score(predictions, y_test_tsfel,average="weighted")
print(f"Precision: {precision}")
```

```python
recall = recall_score(predictions, y_test_tsfel,average="weighted")
print(f"Recall: {recall}")
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(predictions, y_test_tsfel)
print(f"Confusion Matrix:\n{mat}")
```
